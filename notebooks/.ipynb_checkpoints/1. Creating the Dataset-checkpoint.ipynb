```python
# Importing required libraries
import json
import numpy as np
import pandas as pd
import tqdm as tqdm
import seaborn as sns
import matplotlib.pyplot as plt
import re

import urllib.request
from PIL import Image
from io import BytesIO

from pygments import highlight
from pygments.lexers import JsonLexer
from pygments.formatters import TerminalFormatter

from google_play_scraper import Sort, reviews, app
```

```python
app_packages = [
    'com.flipkart.android',
    'com.flipkart.shopsy',
    'in.amazon.mShop.android.shopping',
    'com.myntra.android',
    'com.bigbasket.mobileapp',
    'com.Dominos',
    'com.olx.southasia',
    'com.ril.ajio',
    'com.snapdeal.main',
    'com.jpl.jiomart',
    'com.fsn.nykaa',
    'com.manash.purplle',
    'in.swiggy.android',
    'com.grofers.customerapp',
    'com.zeptoconsumerapp'
]
```

```python
len(app_packages)
```

```python
app_infos = []

for ap in tqdm.tqdm(app_packages):
    info = app(ap, lang='en', country='in')
    app_infos.append(info)
```

```python
def print_json(json_object):
    json_str = json.dumps(json_object, 
                          indent=2, 
                          sort_keys=True,
                          default=str)
    
    print(highlight(json_str, JsonLexer(), TerminalFormatter(style='friendly')))
```

```python
print_json(app_infos[0])
```

```python
# The formatting of the above apps are not ideal
# Function to format the titles

def format_title(title):
    # Use regex to remove everything after separators (like -, :, ,)
    cleaned_title = re.split(r'\s*[-:,|–—•]\s*', title)[0]
    return cleaned_title.strip()
```

```python
fig, axes = plt.subplots(3, 5, figsize=(15, 9))  # 3 rows, 5 columns for 15 apps

# Flatten the axes for easier iteration
axes = axes.flatten()

for i, ax in enumerate(axes):
    if i < len(app_infos):
        ai = app_infos[i]
        title = format_title(ai['title'])
        with urllib.request.urlopen(ai['icon']) as url:
            img = Image.open(BytesIO(url.read()))
        ax.imshow(img)
        ax.set_title(title[:30])  # Trim length to fit nicely
    ax.axis('off')
    
plt.tight_layout()
plt.show()
```

```python
for i, ai in enumerate(app_infos):
    print(f"{i+1}. {format_title(ai['title'])}")
```

```python
for ai in app_infos:
    ai['title'] = format_title(ai['title'])

app_infos_df = pd.DataFrame(app_infos)
```

## Get App Reviews

```python
app_reviews = []

for ap in tqdm.tqdm(app_packages):
    for score in [1,2,4,5]:
        for sort_order in [Sort.MOST_RELEVANT, Sort.NEWEST]:
            try:
                rvs, _ = reviews(
                    ap,
                    lang='en',
                    country='in',
                    sort=sort_order,
                    count=800,
                    filter_score_with=score
                )

                for r in rvs:
                    r['sortOrder'] = 'most relevant' if sort_order == Sort.MOST_RELEVANT else 'newest'
                    r['appID'] = ap
                    r['score'] = score
                app_reviews.extend(rvs)
            except Exception as e:
                print(f"Error for {ap} with score {score}: {e}")
```

```python
print_json(app_reviews[0])
```

```python
app_reviews_df = pd.DataFrame(app_reviews)
```

```python
app_reviews_df.shape
```

```python
app_reviews_df.head()
```

```python
app_reviews_df = app_reviews_df[app_reviews_df['content'].str.len() > 150]
```

```python
app_reviews_df.shape
```

```python

df_1_star = app_reviews_df[app_reviews_df['score'] == 1]
df_2_star = app_reviews_df[app_reviews_df['score'] == 2]
df_4_star = app_reviews_df[app_reviews_df['score'] == 4]
df_5_star = app_reviews_df[app_reviews_df['score'] == 5]
```

```python
print(len(df_1_star), len(df_2_star), len(df_4_star), len(df_5_star))
```

```python
# Sample the required number of reviews for each class
df_1_star_balanced = df_1_star.sample(n=6691, random_state=42)
df_2_star_balanced = df_2_star.sample(n=6691, random_state=42)
df_4_star_balanced = df_4_star.sample(n=6691, random_state=42)
df_5_star_balanced = df_5_star.sample(n=6691, random_state=42)
```

```python
# Concatenate the balanced dataframes
balanced_reviews = pd.concat([df_1_star_balanced, df_2_star_balanced, df_4_star_balanced, df_5_star_balanced])
```

```python
# Shuffle the data
balanced_reviews = balanced_reviews.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
# Verifying the distribution
print(balanced_reviews['score'].value_counts())
```

```python
balanced_reviews.to_csv('../data/reviews.csv',index=None,header=True)
```
